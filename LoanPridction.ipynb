```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("DATA.csv")
```

```python
data.head(10)
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data["Gender"].fillna(data["Gender"].mode()[0],inplace = True)
```

```python
data["Married"].fillna(data["Married"].mode()[0],inplace = True)
```

```python
data["Dependents"] = data["Dependents"].replace("3+",3)
```

```python
data["Dependents"].fillna(data["Dependents"].mode()[0],inplace = True)
```

```python
data["Self_Employed"].fillna(data["Self_Employed"].mode()[0],inplace = True)
```

```python
data["LoanAmount"].fillna(data["LoanAmount"].median(),inplace = True)
```

```python
data["Loan_Amount_Term"].fillna(data["Loan_Amount_Term"].median(),inplace = True)
```

```python
data["Loan_Status"].value_counts()
```

```python
data = pd.get_dummies(data,columns = ["Gender", "Married", "Education", "Self_Employed", "Property_Area","Loan_Status"],drop_first = True)
```

```python
x = data.drop("Loan_Status_Y", axis=1)
y = data["Loan_Status_Y"]
```

```python
x = x.drop("Loan_ID", axis = 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size = 0.2, random_state = 42, stratify = y
)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(
    random_state = 42,
)
dt.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
y_pred = dt.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("Decision Tree \n",classification_report(y_test, y_pred))
```

```python
print("Decision Tree \n",confusion_matrix(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(
    n_estimators = 100,
    random_state = 42,
)

rf.fit(x_train, y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
print("RF Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
print("Random Forest \n",classification_report(y_test, y_pred_rf))
```

```python
print("Random Forest \n",confusion_matrix(y_test, y_pred_rf))
```
